```python
base_url = "https://rategain.com/blog"
total_pages = 45
```

```python
all_extracted_data = []
```

```python
for page_number in range(1, total_pages + 1):
    # Construct the URL for each page
    current_url = f"{base_url}/page/{page_number}" if page_number > 1 else base_url

    # Make a request to the page
    response = requests.get(current_url, headers={'User-Agent': 'Mozilla/5.0'})

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract information from each blog post on the page
        all_blog_posts = soup.select('.blog-item')

        # Iterate over each blog post
        for post in all_blog_posts:
            blog_title = post.select_one('h6 a').text.strip()
            blog_date = post.select_one('.bd-item span').text.strip()
            blog_url = post.select_one('h6 a')['href']
            img_element = post.select_one('.img .rocket-lazyload')
            blog_image_url = img_element['data-bg'] if img_element and 'data-bg' in img_element.attrs else 'No Image'
            likes_count = post.select_one('.zilla-likes span').text.strip()

            # Append the extracted data to the list
            all_extracted_data.append({
                'Blog Title': blog_title,
                'Blog Date': blog_date,
                'Blog URL': blog_url,
                'Blog Image URL': blog_image_url,
                'Likes Count': likes_count
            })
```

```python
from flask import Flask, jsonify, request
from flask_cors import CORS
from fuzzywuzzy import process

app = Flask(__name__)
CORS(app)
# Your data on the server

# Extract titles for fuzzy matching
titles = [entry['Blog Title'] for entry in all_extracted_data]

@app.route('/search', methods=['POST'])
def search():
    # Get the search query from the request
    query = request.json['query']
    print(query)

    # Perform fuzzy matching
    matches = process.extract(query, titles, limit=3)
#     matches = process.extract(query, titles, limit=3)
    # Return the matched entries
    results = [{'Blog Title': title,  'Score': score} for (title, score) in matches]
    titles_only = [result['Blog Title'] for result in results]
    
    filtered_entries = [
    {
        'Blog Title': entry['Blog Title'],
        'Blog Image URL': entry['Blog Image URL'],
        'Blog URL': entry['Blog URL']
    }
    for entry in all_extracted_data if entry['Blog Title'] in titles_only
    ]
    
    formatted_results = []

    for entry in filtered_entries:
        formatted_entry = {
            'Blog Title': entry['Blog Title'],
            'Blog Image URL': entry['Blog Image URL'],
            'Blog URL': entry['Blog URL'],
        }
        formatted_results.append(formatted_entry)
    for entry in filtered_entries:
        print("Blog Title:", entry['Blog Title'])
        print("Blog Image URL:", entry['Blog Image URL'])
        print("Blog URL:", entry['Blog URL'])
        print("\n" + "-"*30 + "\n") 

    

#     results = [{'Blog Title': title, 'Blog Image URL': next((entry['Blog Image URL'] for entry in data if entry['Blog Title'] == title), 'N/A'), 'Blog URL': next((entry['Blog URL'] for entry in data if entry['Blog Title'] == title), 'N/A'), 'Score': score} for (title, score) in matches]

    return jsonify(filtered_entries)

if __name__ == '__main__':
    app.run(debug = False)
```
